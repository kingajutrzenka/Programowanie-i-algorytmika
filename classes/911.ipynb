funkcję w programie, który wszystkie imiona żeńskie z tego pliku umieści w pliku
panie.txt, a wszystkie imiona męskie umieści w pliku panowie.txt. Imiona powinny
być oddzielone przecinkiem i znakiem spacji oraz w kolejności zgodnej z kolejnością w pliku
imiona.txt.

Dodatkowo, w ostatnim wierszu każdego pliku wynikowego zapisz liczbę umieszczonych w
nim imion.

```python
def czy_zenskie(imie):
  return imie[-1] == 'a'

czy_zenskie("Anna"), czy_zenskie("Kamil")
```

```python
with open('imiona.txt', 'r', encoding="utf-8") as plik:
  for linia in plik:
    lista= linia.strip().split()
    print(lista)
```

```python
iz = 0
im = 0
with open('panie.txt', 'w', encoding="utf-8") as panie, open('panowie.txt', 'w', encoding="utf-8") as panowie:
  for el in lista:
    if czy_zenskie(el):
      panie.write(el + ", ")
      iz += 1
    else:
      panowie.write(el + ", ")
      im += 1
  panie.write("\n" + str(iz))
  panowie.write("\n" + str(im))
```
