```python
def newton(n, k):
  if k == 0 or k == n:
    return 1
  return newton(n-1, k-1) + newton(n-1, k)
newton(4, 2)
```

```python
def NWD(a, b):
  if b == 0 and a != 0:
    return a
  elif b != 0 and a == 0:
    return b
  return NWD(b, a%b)
```

```python
def Fib(m):
  if m == 0:
    return 0
  elif m == 1:
    return 1
  return Fib(m-1) + Fib(m-2)
```

```python
def Fib2(m):
  if m == 0:
    return "b"
  if m == 1:
    return "a"
  return Fib2(m-1) + Fib2(m-2)
```

### **4**

```python
def Fib3(m):
  if m == 0:
    return "0"
  if m == 1:
    return "01"
  return Fib3(m-1) + Fib3(m-2)
```

### **Uniwersalna**

```python
def FibAA(m, a, b):
  if m == 0:
    return a
  if m == 1:
    return b
  return FibAA(m-1, a, b) + FibAA(m-2, a, b)
```

```python
FibAA(5, 0, 1), FibAA(5, 2, 1), FibAA(5, 'b', 'a'), FibAA(5, '0', '01')
```
