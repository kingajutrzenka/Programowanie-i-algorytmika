```python
#warunek
import math as m
czy_w_kółku = 0
for i in range(n):
  a = random.random()
  b = random.random()
  if m.sqrt(a**2 + b**2) <= 1:
    czy_w_kółku += 1
print((czy_w_kółku*4)/n)

```

## **Zadanie 5.**

Dane: funkcja f(x) ciągła w przedziale domkniętym [a, b] i spełniająca f(a)*f(b), epsilon - długość przedziału w którym znajduje się miejsce zerowe.

Wynik: Przybliżenie x* zera funkcji w przedziale [a, b] z oczekiwana dokładnością epsilon

image.png

```python
a = -50
b = 10
epsilon = 1

def f(x):
    return (x**2)

while abs(a - b) > epsilon:
    x = (a + b) / 2
    if f(x) == 0:
        break
    if f(a) * f(x) < 0:
        b = x
    else:
        a = x

x = (a + b) / 2
print(x)
```
