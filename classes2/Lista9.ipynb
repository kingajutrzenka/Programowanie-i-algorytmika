Sporządź arkusz kalkulacyjny dla zachłannej metody wydawania reszty w złotówkach za pomocą
najmniejszej liczby banknotów i monet. Napisz również program w języku Python rozwiązujący ten

```python
a = 1328
g = [500, 200, 100, 50, 20, 10, 5, 2, 1]
while a != 0:
  for i in g:
    if a >= i:
      print(i)
      a -= i
      break

```
