```python
def det(p, q, r): #bardzo ważno ważna
  xp, yp = p
  xq, yq = q
  xr, yr = r
  return xr*(yp - yq) - xq*(yp - yr) + xp*(yq - yr)

def sgn(a):
  if a>0:
    return 1
  if a<0:
    return -1
  return 0

p = (1,1)
q = (10, 1)
r = (6, 5)

print(det(p, q, r))
print(sgn(det(p, q, r)))
```

```python
def det(p, q, r): #bardzo ważno ważna
  xp, yp = p
  xq, yq = q
  xr, yr = r
  return xr*(yp - yq) - xq*(yp - yr) + xp*(yq - yr)

def sgn(a):
  if a>0:
    return 1
  if a<0:
    return -1
  return 0

p = (1,1)
q = (10, 1)
r = (5, 1)

x=(sgn(det(p, q, r)))

xp, yp = p
xq, yq = q
xr, yr = r

if x==0:
  if xp <= xr <= xq and yp <= yr <= yq and x==0:
    print("punkt należy do odcinka")
  else:
    print("punkt nie należy do odcinka")

if det(p, q, r) >0 :
  print({"po lewej"})
elif det(p, q, r) == 0 :
  print("jest współliniowy")
else:
  print("po prawej")
print(det(p, q, r))
print(sgn(det(p, q, r)))
```
