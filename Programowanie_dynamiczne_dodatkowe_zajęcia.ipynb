**Programowanie dynamiczne (PD)** to technika algorytmiczna służąca do rozwiązywania złożonych problemów przez rozkładanie ich na mniejsze, zachodzące na siebie podproblemy, zapamiętywanie wyników tych podproblemów (memoizacja) i łączenie ich w celu znalezienia optymalnego rozwiązania. Zamiast wielokrotnie obliczać te same podproblemy, PD przechowuje ich rozwiązania w pamięci (np. w tablicy), co znacznie przyspiesza działanie algorytmu, zamieniając czas na pamięć. Jest to kluczowe dla optymalizacji problemów rekurencyjnych, które mają "optymalną podstrukturę" i "overlapping subproblems" (nakładające się podproblemy).

## **Najdłuższy wspólny podciąg:**

L **A** T **O**

S **A** M **O**

to 2, nie muszą być literki koło siebie, ale ważna jest kolejność


```python
a = "LATO"
b = "SAMOLOTA"
#n - wiersze
#m - kolumny
def np(w1, w2):
  print(w1, w2)
  n = len(w1)
  m = len(w2)
  print(n), print(m)
  tabelka = []
  for i in range(n+1):
    tabelka.append([0 for j in range(m+1)]) #mniej bezpieczna opcja: [0]*(m+1)
  for k in range(1, n+1):
    for q in range(1, m+1):
      if w1[k-1] == w2[q-1]: # Poprawiony błąd: zamieniono indeksy
         tabelka[k][q] = tabelka[k-1][q-1] + 1
      else:
        tabelka[k][q] = max(tabelka[k-1][q], tabelka[k][q-1])

  for w in tabelka: #jest to lepsze wypisywanie, niż print(tabelka)
    print(w)
  return tabelka[-1][-1]
np(a, b)
```

## **Backtracking**
przykład funkcji rekurencyjnej

szukanie hetmanów

```python
długość = 8
def hetmani(n):
  h = [-1 for i in range(n)]

  def sprawdz(het, idx):
    if idx == n:
      print(het) # Zmieniono 'h' na 'het' aby uniknąć UnboundLocalError
      return #1

    for i in range(n):
      h = het[:] # kopia
      szach = False # Inicjalizacja zmiennej szach, aby uniknąć UnboundLocalError

      for k in h[:idx]:
        if k == 1:
          szach = True
        if szach:
          continue

      for ki in range(idx):
        k = h[ki]
        if abs(idx - ki) == abs(i - k):
          szach = True
      if szach:
        continue

      h[idx] = i
      sprawdz(h, idx + 1)


  sprawdz(h, 0)


hetmani(długość)
```
