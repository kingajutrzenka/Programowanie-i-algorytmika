Dany jest plik tekstowy imiona.txt, w którym znajdują się imiona żeńskie wpisane po  spacji. Napisz funkcję sprawdzająca, czy dane imię jest imieniem żeńskim. Wykorzystaj tę funkcję w programie, który wszystkie imiona żeńskie z tego pliku umieści w pliku panie.txt, a wszystkie imiona męskie umieści w pliku panowie.txt. Imiona powinny być oddzielone przecinkiem i znakiem spacji oraz w kolejności zgodnej z kolejnością w pliku imiona.txt.

Dodatkowo, w ostatnim wierszu każdego pliku wynikowego zapisz liczbę umieszczonych w nim imion.

**Odczytywanie plików**

```python
with open ('imiona.txt', 'r') as wodka:
  for Janusz in wodka:
    print(Janusz)
```

```python
def czy_zenskie(imie):
  return imie[-1] == 'a'

czy_zenskie("Anna"), czy_zenskie("Kamil")
```

```python
with open ('imiona.txt', 'r') as wodka:
  for Janusz in wodka:
    alko = Janusz.strip().split() #WAŻNA KOLEJNOŚĆ!!! Strip -> split
    print(alko)
```

```python
iz = 0
im = 0
with open ('panie.txt', 'w') as panie, open ('panowie.txt', 'w') as panowie: #bo tworzymy nowy plik
 for drink in alko:
    if czy_zenskie(drink):
      panie.write(drink + ", ")
      iz += 1
    else:
      panowie.write(drink + ", ")
      im += 1
 panie.write("\n" + str(iz)) # "\n - wykonaj w nowej linijce",  STR - zawsze w pliku zapisujemy wszystko jako str!!!!
 panowie.write("\n" + str(im))
```
